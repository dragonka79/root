Supervised learning works: Training data-set &rarr; feed this data-set to the learning algorithm &rarr; it gives an __h__ function known as hypothesis to make predictions.<br>
For the example above the input for the hypothesis is the _size_ of a house and the output is the estimated _price_ of the house.<br>
The job of the learning algorithm is to __produce a hypothesis__ based on the training set as input to predict the price of other houses based on their size.<br><br>
!Note that making predictions on one input value only cannot give a good prediction in general, now for better understanding when examples need to be shown; we were talking about more about it later on within this paragraph. <br><br>

#### The Theory
When designing a learning algorithm the question is how to represent $h$.<br>
In case when the learning algorithm is linear regression the hypothesis can be expressed as:<br>
<br>
$$h(x)=\sum_{j=0}^{m} \theta_{j}x_{j}$$
where $x_{0} = 1$, $x_{j}$ is the input features, $m$ = number of training examples, $n$= number of features and $\theta$ means __parameters__ of the learning algorithm.<br>
$\theta = \begin{bmatrix} \theta_{0} \\ \theta_{1} \\ \theta_{2} \\ \vdots \\ \theta_{n}\end{bmatrix} \qquad x = \begin{bmatrix} x_0 \\ x_1 \\ x_2 \\ \vdots \\ x_{n}\end{bmatrix}$ 
<br>
(x,y) = training example and let $(x^{(i)},y^{(i)}) = i^{th}$ training example, $i=1,..,m$ <br>
The job of the learning algorithm is to choose parameters $\theta$. $\theta$ means the slope and the intercept when $n=1.$<br><br>
So in case more input (input feature) is added, say the number of the bedrooms, then the input features and the hypothesis are as follows based on the above example:<br>
$x_{1}$ = the size of the house <br>
$x_{2}$ = the number of the bedrooms <br>
$n = 2$ <br>
$m = 4$ <br>
$x_1^{(1)}=2104 \quad x_1^{(2)}=1416$ <br>
$h(x)=\theta_{0} + \theta_{1}x_{1} + \theta_{2}x_{2}$
<br><br>

#### How to choose parameters $\theta$ ?<br>
Choose $\theta$ such to $h(x)\approx y$ for the training examples.<br>
$h(x)$ depends both $\theta$ and on the input features $x$, therefore the hypothesis can be written as: $h_{\textstyle \theta}(x)$. <br>
In the linear regression algorithm we would like to minimize the squared differences (squared residuals) between the hypothesis outputs and the correct output values. __The smaller the total squared difference, the closer runs the hypothesis to the true values in total__, so our goal is:  <br>
$J(\theta$) = $\underset{\textstyle \theta}{\textrm{minimize}} \frac{1}{2} \sum_{j=0}^{m}(h_{\textstyle \theta}(x^{(i)}) - y^{(i)})^2 $, &emsp; where the $\frac{1}{2}$ agreed upon constant multiplier. Note that the squared residuals is a type of loss function.<br><br>
For linear regression, the most common loss function used is the Mean Squared Error (MSE). The MSE calculates the average of the squared differences between the predicted values and the actual target values:<br>
MSE = $\frac{1}{m} \sum_{j=0}^{m}(h_{\textstyle \theta}(x^{(i)}) - y^{(i)})^2$

#### How to implement an algorithm to find $\theta$ that minimizes $J(\theta)$?
Gradient descent algorithm:
 1. Take the derivative of the square difference for each parameter in it. Normalize the input data if needed. Using different terminology take the gradients of the loss function.
 2. Start with some $\theta$, it can be anything, say $\theta = \vec 0$.<br>
 3. Keep changing $\theta$ to reduce $J(\theta)$.

Formalizing the gradient descent algorithm:
$\theta_{j} := \theta_{j} - \alpha \sum_{i=0}^{m} \frac{\textstyle{\partial}}{\textstyle{\partial \theta_{j}}}J(\theta)$,&emsp; where $\alpha$ is the learning rate.<br>
$\frac{\textstyle{\partial}}{\textstyle{\partial \theta_{j}}}J(\theta) = ... = \sum_{j=0}^{m}(h_{\textstyle \theta}(x^{(i)}) - y^{(i)})x_{j}^{(i)}$<br>
$\alpha \sum_{i=0}^{m} \frac{\textstyle{\partial}}{\textstyle{\partial \theta_{j}}}J(\theta)$ is the step size of approaching the optimal $\theta_{j}$ value.<br>
The gradient descent algorithm can stop when the step size is very close to $0$, in practice $0.001$ or smaller.<br>
!Note: the learning rate is very sensitive to the input features, when the input features have large ranges. It means when we tries to make gradient descent work on the given input features as-is and $\alpha=0.01$, after the second interation the moving step turns out to be too big and the line diverges.<br>
The solution can be two-fold: either setting the learning rate to a smaller value from the second iteration, say $10^{-8}$ or normalizing the input. <br>


#### Input Normalization
Normalization scales input features to a smaller, more consistent range — usually around 0 or 1 — to make training more stable and allow larger learning rates.<br>
There are two types of input features normalization methods:
1. Min-Max Normalization: $x_{norm} = \frac{\textstyle{x−min(x)}}{\textstyle{max(x)−min(x)}}$<br>
2. Z-score Normalization (Standardization). Scales data to have mean 0 and std deviation 1:
   $x_{norm} =\frac{\textstyle{x-\mu}}{\textstyle{\sigma}}$, &emsp; where $\textstyle{\mu}$ is the mean of x and $\textstyle{\sigma}$ is the standard deviation. <br>

## Let's see how gradient descent works for the example above
Since $n=1$, it means we have $1$ input feature only, so for the sake of simplicity $x_1:=x$<br><br>
$x^{(1)}=2104,\quad y^{(1)}=400$<br>
$x^{(2)}=1416,\quad y^{(2)}=232$<br>
$x^{(3)}=1534,\quad y^{(3)}=315$<br>
$x^{(4)}=852,\quad y^{(4)}=178$<br>
$x_{0}=1$ <br>
Let us choose $\theta(b,m) = (0,0)$ and $\alpha = 0.01$, &emsp; where $b$=the intercept of $h$ and $m$=slope of $h$.<br>
Therefore $h=0$<br>
We have now the basic data and the starting parameters choosen, so let's get started with the first step of gradient descent algorithm.



#### Step 1.0: Normalize the input with Min-Max Normalization

- $min(x)=852$
- $max(x)=2104$
- $range=1252$

Now normalize:<br>
$x^{(1)}$&rarr;$x_{norm}^{(1)}$:&emsp;$2104$&rarr;$\frac{2104-852}{1252}=1.0$<br><br>
$x^{(2)}$&rarr;$x_{norm}^{(2)}$:&emsp;$1416$&rarr;$\frac{1416-852}{1252}\approx 0.45$<br><br>
$x^{(3)}$&rarr;$x_{norm}^{(3)}$:&emsp;$1534$&rarr;$\frac{1534-852}{1252}\approx 0.545$<br><br>
$x^{(4)}$&rarr;$x_{norm}^{(4)}$:&emsp;$852$&rarr;$\frac{852-852}{1252}=0$<br><br>

#### Step 1.2: Compute Gradients
$\frac{\textstyle{\partial}}{\textstyle{\partial{m}}}$ = $\frac{2}{\textstyle{m}}\sum e^{(i)}$ · $x^{(i)}$ = $\frac{2}{4}$ · $(-676.075)$ = $-338.038$<br>
$\frac{\textstyle{\partial}}{\textstyle{\partial{b}}}$ = $\frac{2}{\textstyle{m}}\sum e^{(i)}$ = $\frac{2}{4}$ · $(-1125)$ = $-562.5$<br>

#### Step 1.3: Update parameters

$m:=0-0.01 ·(-338.038) = 3.380$<br>
$b:=0-0.01 ·(-562.5) = 5.625$<br>
